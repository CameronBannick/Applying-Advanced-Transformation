# Imports

```python
import numpy as np
import pandas as pd
import os,json
```

# Importing and Cleaning First Dataframe

```python
#import first dataframe
info_df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vS1ZstYLwFgwhZnqDsPjtnlHYhJp_cmW55J8JD5mym0seRsaem3px7QBtuFF0LiI7z1PLCkVKAkdO7J/pub?output=csv')
info_df.head()
```

```python
#check for duplicates/missing values
print(f'Superhero Info DF contains {info_df.duplicated().sum()} duplicates')
print(f'Superhero Info DF contains {info_df.isna().sum().sum()} Missing Values')
```

```python
#checking info on info_df
info_df.info()
```

```python
#splitting Hero|Publsiher column into two seperate columns
info_df['Hero|Publisher'].head(2)
```

```python
info_df[['Hero', 'Publisher']] = info_df['Hero|Publisher'].str.split('|', expand=True)
info_df.head(2)
```

```python
#dropping old Hero|Publisher column
info_df = info_df.drop(columns='Hero|Publisher')
info_df.head(2)
```

```python
#inspecting Measurements dictionary
measr = info_df.loc[0, 'Measurements']
print(f'Measurements Column is a {type(measr)} type')
measr
```

Since there are single quotes present, we will need them to replaced with double quotes in order to get JSON to work with the data

```python
#adjusting quotes to ensure data readable by JSON
info_df['Measurements'] = info_df['Measurements'].str.replace("'", '"')
info_df['Measurements'] = info_df['Measurements'].apply(json.loads)
info_df['Measurements'].head()
```

```python
#test to ensure success in previous cell
test = info_df.loc[0, 'Measurements']
print(f'Measurement Column is now a {type(test)} type')
test
```

```python
#creating a series based on Measurements Dictionary
hite_wate = info_df['Measurements'].apply(pd.Series)
hite_wate
```

```python
#combining series with info dataframe
info_df = pd.concat((info_df, hite_wate), axis=1)
info_df.head(2)
```

```python
#dropping now redundant Measurements Column
info_df = info_df.drop(columns='Measurements')
info_df.head()
```

```python
info_df['Height'] = info_df['Height'].str.replace(' cm', '')
```

```python
info_df.head()
```

```python
info_df['Weight'] = info_df['Weight'].str.replace(' kg', '')
info_df.head()
```

```python
info_df['Height'] = pd.to_numeric(info_df['Height'])
info_df['Weight']= pd.to_numeric(info_df['Weight'])
info_df.info()
```

# Importing and Cleaning Second Dataset

```python
#importing power_df
power_df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vSzdWOBaXOoz52vPmCFV5idNlDBohLY1Lsbc1IfZIZQ7cV_aNB2wYBfhF49uE1TaO1B5MQCGWiNrFfd/pub?output=csv')
power_df.head()
```

```python
#check for duplicates/missing values
print(f'Superhero Power DF contains {power_df.duplicated().sum()} duplicates')
print(f'Superhero Power DF contains {power_df.isna().sum().sum()} Missing Values')
```

```python
#inspecting info
power_df.info()
```

```python
#inspecting powers column
power = power_df.loc[0, 'Powers']
print(f'The Powers Column is a {type(power)} type')
power
```

All powers are labeled in a single string. Will use expand to fix

```python
#turn to list 
power_df['Power_Split']=power_df['Powers'].str.split(',', expand=False)
power_df.head()
```

```python
#inspecting powers column
power = power_df.loc[0, 'Power_Split']
print(f'The Powers Column is a {type(power)} type')
power
```

```python
power_df['Power_Split'].value_counts()
```

```python
#sepearting powers from list
exploded = power_df.explode('Power_Split')
exploded.head()
```

```python
#cols_to_make variable for loop
cols_to_make = exploded['Power_Split'].dropna().unique()
cols_to_make
```

```python
#loop for hot encoding
for col in cols_to_make:
    power_df[col]=power_df['Powers'].str.contains(col)
#inspect results
power_df.head()
```

```python
#dropping Powers and Powers_Spilt Column
power_df = power_df.drop(columns=['Powers','Power_Split'])
power_df.head()
```

# Combining Dataframes

```python
#combine both dfs
hero_df = pd.merge(info_df, power_df, left_on='Hero', right_on='hero_names')
hero_df.head()
```
